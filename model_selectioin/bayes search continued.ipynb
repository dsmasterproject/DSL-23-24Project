```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import csv
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse 
from skopt import BayesSearchCV
```

```python
development = pd.read_csv(r'C:\Users\feder\Desktop\POLI\DSL\DSL_Winter_Project_2024\DSL_Winter_Project_2024\development.csv')
evaluation = pd.read_csv(r'C:\Users\feder\Desktop\POLI\DSL\DSL_Winter_Project_2024\DSL_Winter_Project_2024\evaluation.csv')
```

```python
df = pd.concat([development, evaluation], sort=False)
```

```python
df.set_index('Id', inplace=True)
```

```python
outliers_idx = [0,7,12,15,16,17]
dropped_cols = ['pmax[%s]' %i for i in outliers_idx]+['negpmax[%s]' %i for i in outliers_idx]+['area[%s]' %i for i in outliers_idx]+['tmax[%s]' %i for i in outliers_idx]+['rms[%s]' %i for i in outliers_idx]
print(dropped_cols)
```

```python
df = df.drop(columns=dropped_cols)
display(df)
```

```python
coord=['x', 'y']
test_mask = df['x'].isna() & df['y'].isna()

X = df.drop(columns=coord).values
y = df[coord].values

X_train_valid = X[~test_mask]
y_train_valid = y[~test_mask]

X_test = X[test_mask]
y_test = y[test_mask]

from sklearn.model_selection import train_test_split
rs = 2024

X_train, X_valid, y_train, y_valid = train_test_split(X_train_valid,y_train_valid,shuffle=True, random_state=rs)
```

```python
parameters = {"n_estimators":[100,250], "max_depth":[50,100], "min_samples_split": [2,5,10], "max_features":[ "sqrt", "log2"], "random_state":[rs], "n_jobs":[-1] } 
bayes = BayesSearchCV(RandomForestRegressor(), search_spaces=parameters, scoring="r2", cv=5, n_iter=50, n_jobs=-1, random_state=rs)
```

```python
np.int = int
bayes.fit(X_train_valid, y_train_valid)
```

```python
bayes.best_params_
```

Inserisci sotto i valori dei parametri che hai ottenuto sopra

```python
rfr = RandomForestRegressor(n_estimators= 100,min_samples_split= 5,max_features='sqrt',max_depth= 100 ,random_state=rs, n_jobs= -1)
rfr.fit(X_train, y_train)
y_rfr=rfr.predict(X_valid)
distance=np.mean(np.sqrt(np.sum((y_rfr - y_valid) ** 2, axis=1)))
print(distance)
evaluation = evaluation.drop(columns=dropped_cols)
eva_df = evaluation.drop(columns=['Id'])
eva_y=rfr.predict(eva_df)

df = pd.DataFrame(eva_y, columns=['Id', 'Predicted'])
df['Predicted'] = df.apply(lambda row: f"{row['Id']:.1f}|{row['Predicted']:.1f}", axis=1)
df['Id'] = df.index
df.to_csv('outputbayes2.csv', index=False, sep=',')

```
